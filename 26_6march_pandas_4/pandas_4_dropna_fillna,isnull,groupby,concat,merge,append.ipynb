# merge(on= ,how=) /merge(left_on= ,right_on=)

```python
employee=[(1,'name1'),(3,'name3'),(2,'name2')]

dept=[(2,'science','bbsr'),(1,'arts','angul'),(5,'eco','bari')]

df1=pd.DataFrame(employee,columns=['empid','name'])
df1


```

```python
df2=pd.DataFrame(dept,columns=['empid','course','location'])
df2
```

```python
df3=pd.merge(df1,df2,on='empid')

df3
```

```python
#left merge
```

```python
pd.merge(df1,df2,on='empid',how='left')
```

```python
#right merge
```

```python
pd.merge(df1,df2,on='empid',how='right')
```

```python
#outer
```

```python
pd.merge(df1,df2,on='empid',how='outer')
```

```python
employee=[(1,'name1'),(3,'name3'),(2,'name2')]

dept=[(2,'science','bbsr'),(1,'arts','angul'),(5,'eco','bari')]

df1=pd.DataFrame(employee,columns=['empid','name'])
df2=pd.DataFrame(dept,columns=['id','course','location'])

```

```python
df1
```

```python
df2
```

```python
df3=pd.merge(df1,df2,left_on=['empid'],right_on=['id'])

df3
```
